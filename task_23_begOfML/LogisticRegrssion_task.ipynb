```python
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as pltt
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df
```

```python
df.isna().value_counts()
```

```python
df.groupby('left').mean()
```

```python
dff =df[['satisfaction_level','promotion_last_5years','average_montly_hours']]
dff
```

```python
dum = pd.get_dummies(df.salary)
dum
```

```python
inpu = pd.concat([dff,dum],axis='columns')
inpu
```

```python
out = df.left
out
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
inpu_train , inpu_test , out_train , out_test = train_test_split(inpu,out,train_size=0.8)
```

```python
model =LogisticRegression()
model.fit(inpu_train,out_train)
```

```python
model.predict(inpu_test)
```

```python
model.score(inpu_train,out_train)
```

```python
model.score(inpu_test,out_test)
```
