# Task: make the accuracy 98% or more

```python
import pandas as pd 
df = pd.read_csv('heart.csv')
df
```

```python
target = df['target']
target
```

```python
df = df.drop('target',axis='columns')
df
```

# ranges need to be normalized

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = df.columns.to_list()
for one in range(len(cols)):
    df[cols[one]] = scaler.fit_transform(df[[cols[one]]])

df.head()
```

# in order to find the best model with best accuracy grid search tool will be used


```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
```

```python
params = {'Lreg': {'model':LogisticRegression(),'params':{}},
         'svm': {'model':svm.SVC(),'params':{'C' :[6,5,3],'kernel':['rbf','linear']}},
         'Dtree':{ 'model':DecisionTreeClassifier(),'params':{}},
         'Rforest':{'model':RandomForestClassifier(),'params':{'n_estimators':[10,40,50,100]}},
         'NB': {'model':GaussianNB(),'params':{}}}
scores =[]
for key , value in params.items():
    Gsearch = GridSearchCV(value['model'],value['params'],return_train_score=False)
    Gsearch.fit(df,target)
    scores.append({'model_name': key, 'best_score':Gsearch.best_score_ , 'best_params': Gsearch.best_params_})
df_final = pd.DataFrame(scores)
df_final
    
```

# now let's try KFold

```python

def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test) # c

from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=20)

scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(df,target):
   
    X_train, X_test, y_train, y_test = df.iloc[train_index], df.iloc[test_index], target.iloc[train_index], target.iloc[test_index]
    scores_logistic.append(get_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X_train, X_test, y_train, y_test))  
    scores_svm.append(get_score(svm.SVC(), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
print(f'Lreg scores: {scores_logistic}')
print(f'svm scores : {scores_svm} ')
print(f'Random forest scores : {scores_rf}')      
    
```
