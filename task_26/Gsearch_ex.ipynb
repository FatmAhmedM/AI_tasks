```python
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd 
from sklearn import svm
digits = load_digits()
```

```python
multi_parameters = {'svm': { 'model':svm.SVC(), 'params': {'C':[2,6,15],'kernel' : ['rbf','linear']}},
                    'Lreg': {'model':LogisticRegression(), 'params': {'C':[2,6,15]}},
                    'Dtree': {'model':DecisionTreeClassifier(),'params': {}},
                    'NB': {'model': GaussianNB(),'params':{} },
                    'Rforest':{'model':RandomForestClassifier(),'params': {'n_estimators':[1,5,10]}}

                    
    
}
```

```python
score_OFmodels = []
for key , value in multi_parameters.items():
    Gsearch = GridSearchCV(value['model'],value['params'],cv= 2,return_train_score=False)
    Gsearch.fit(digits.data,digits.target)
    score_OFmodels.append({'model_name': key, 'best_score':Gsearch.best_score_ , 'best_params': Gsearch.best_params_})
df = pd.DataFrame(score_OFmodels)
df
```

best one for this case and these params is svm : c =2 and kerenl = rbf
