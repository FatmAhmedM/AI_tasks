```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
mod = SVC(kernel = 'rbf')
mod
```

```python

inpu = df.drop('target',axis='columns')
out = df.target
inpu_train , inpu_test , out_train ,out_test = train_test_split(inpu,out,test_size=0.2,random_state=12)
```
